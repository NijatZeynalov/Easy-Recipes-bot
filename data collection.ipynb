```python
#data preparation
import requests
from bs4 import BeautifulSoup
import time
final_result = []
for i in range(1,5):
    url = 'http://dadli.az/pehriz-yemekleri/page/'+str(i)+'/'
    page = requests.get(url,verify = False)
    soup = BeautifulSoup(page.content, 'html.parser')
    #d = soup.find_all(class_="yemekSiyahiXeber")
    urls = []
    for link in soup.find_all('a'): 
        urls.append(link.get('href')) 
        
    result = [i for i in urls if i.startswith('http://dadli.az/pehriz-yemekleri/') and len(i)>40 and 'video' not in i]
    hello = [x.strip(' ') for x in result]
    final_list = list(set(hello))
    #final_result = []
    final_result.append(final_list)
      #  print(final_list)
    print('page ' + str(i) +' done')
    time.sleep(2)
    
#butun urlleri bir liste yig
import itertools
#a = [['a','b'], ['c']]
final_urls = list(itertools.chain.from_iterable(final_result))
print('Butun yemeklerin sayi: ' + str(len(final_urls)))
```

```python
ad_list = []
terkibi_list = []
hazirlanmasi_list = []
img_url_list = []

for u in final_urls:
    url = u
    page = requests.get(url,verify = False)
    soup = BeautifulSoup(page.content, 'html.parser')

    d = soup.find(class_="fullstory").text
    try:
        terkibi = d[d.find('Tərkibi:') + len('Tərkibi:') : d.find('Hazırlanması')]
        hazirlanmasi = d[d.find('Hazırlanması:') + len('Hazırlanması:') : d.find('Etiketlər')]
    except:
        print(f'Error: {u}')
        continue

    terkibi = ''.join([i for i in terkibi if not i.isdigit()])
    terkibi=terkibi.replace('qr', '')
    terkibi=terkibi.replace('ədəd', '')
    terkibi=terkibi.replace('çay qaşığı', '')
    terkibi=terkibi.replace('xörək qaşığı', '')
    terkibi=terkibi.replace('çay stəkanı', '')
    terkibi=terkibi.replace('kg', '')
    terkibi=terkibi.replace('su stəkanı', '') 
    terkibi=terkibi.replace('yemək qaşığı', '') 
    terkibi=terkibi.replace('½', '')
    terkibi=terkibi.replace('stəkan yarım', '')

    d = soup.find(class_="fullsekil")

    img_url = 'http://dadli.az' + d.find('img').attrs['src']

    d = soup.find(class_="icpanelBasliq")
    ad = d.text.strip()
    ad = ad.replace('resept', '')
    ad = ad.replace(' i', '')
    
    ad_list.append(ad)
    terkibi_list.append(terkibi)
    hazirlanmasi_list.append(hazirlanmasi)
    img_url_list.append(img_url)
    time.sleep(2)
    print('done for ' + ad)
```
